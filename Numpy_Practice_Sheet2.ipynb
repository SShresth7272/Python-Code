```python
import numpy as np

arr = np.array([-1, 0.5, 2.7])
print(np.abs(arr))   # Absolute value
print(np.sin(arr))   # Sine
print(np.exp(arr))   # Exponential
print(np.log(arr+2)) # Natural log (avoid log(0))
```

```python
arr = np.array([90, 180, 560, 456])
print(np.deg2rad(arr))  # Degrees to radians
print(np.rad2deg(arr))  # Radians to degrees
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load an image

try:
    image_arr = plt.imread('/MainBefore.jpg')
except FileNotFoundError:
    # Create a sample image if file doesn't exist
    print("Image file not found. Creating a sample image for demonstration.")
    sample_image = np.random.randint(0, 255, (300, 400, 3), dtype=np.uint8)
    image_arr = sample_image / 255.0  # Normalize to 0-1 range like real images

# Step 2: Explore shape and dtype
print("Original image shape:", image_arr.shape)
print("Original image dtype:", image_arr.dtype)

# Step 3: Convert to grayscale
gray_image = np.mean(image_arr, axis=2)
print("Grayscale image shape:", gray_image.shape)

# Step 4: Crop the image
# Adjust slice indices based on image dimensions
height, width = gray_image.shape
crop_height = min(150, height-50)  # Ensure don't exceed boundaries
crop_width = min(200, width-100)
cropped = gray_image[50:50+crop_height, 100:100+crop_width]
print("Cropped image shape:", cropped.shape)

# Step 5: Apply invert filter
# Handle both normalized (0-1) and 0-255 range images
if gray_image.max() <= 1.0:
    inverted = 1.0 - gray_image
else:
    inverted = 255 - gray_image

# Step 6: Display results
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Original image (if it's color)
if len(image_arr.shape) == 3:
    axes[0, 0].imshow(image_arr)
    axes[0, 0].set_title('Original Color Image')
    axes[0, 0].axis('off')
else:
    axes[0, 0].imshow(image_arr, cmap='gray')
    axes[0, 0].set_title('Original Grayscale Image')
    axes[0, 0].axis('off')

# Grayscale image
axes[0, 1].imshow(gray_image, cmap='gray')
axes[0, 1].set_title('Grayscale Image')
axes[0, 1].axis('off')

# Cropped image
axes[1, 0].imshow(cropped, cmap='gray')
axes[1, 0].set_title('Cropped Image')
axes[1, 0].axis('off')

# Inverted image
axes[1, 1].imshow(inverted, cmap='gray')
axes[1, 1].set_title('Inverted Image')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

# Additional exploration
print("\n--- Image Statistics ---")
print(f"Grayscale image range: {gray_image.min():.3f} to {gray_image.max():.3f}")
print(f"Cropped image range: {cropped.min():.3f} to {cropped.max():.3f}")
print(f"Inverted image range: {inverted.min():.3f} to {inverted.max():.3f}")
```
