```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Define food items and their categories
food_items = {
    'Apple': 'Fruit',
    'Banana': 'Fruit',
    'Orange': 'Fruit',
    'Strawberry': 'Fruit',
    'Grapes': 'Fruit',
    'Bread': 'Bakery',
    'Croissant': 'Bakery',
    'Bagel': 'Bakery',
    'Muffin': 'Bakery',
    'Milk': 'Dairy',
    'Cheese': 'Dairy',
    'Yogurt': 'Dairy',
    'Butter': 'Dairy',
    'Chicken': 'Meat',
    'Beef': 'Meat',
    'Pork': 'Meat',
    'Salmon': 'Seafood',
    'Shrimp': 'Seafood',
    'Rice': 'Grains',
    'Pasta': 'Grains',
    'Cereal': 'Grains'
}

# Generate dates (last 365 days)
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 1, 1)
dates = [start_date + timedelta(days=x) for x in range(0, (end_date - start_date).days)]

# Generate 2000 rows of data
data = []
for _ in range(2000):
    date = np.random.choice(dates)
    item = np.random.choice(list(food_items.keys()))
    category = food_items[item]

    # Generate realistic sales based on category
    if category == 'Fruit':
        sales = np.random.randint(50, 300)
    elif category == 'Bakery':
        sales = np.random.randint(30, 150)
    elif category == 'Dairy':
        sales = np.random.randint(20, 200)
    elif category == 'Meat':
        sales = np.random.randint(80, 400)
    elif category == 'Seafood':
        sales = np.random.randint(100, 500)
    else:  # Grains
        sales = np.random.randint(40, 250)

    data.append({
        'Date': date,
        'Item': item,
        'Category': category,
        'Sales': sales
    })

# Create DataFrame
df = pd.DataFrame(data)

# Sort by date
df = df.sort_values('Date').reset_index(drop=True)

# Display info about the DataFrame
print("Food Sales DataFrame (2000 rows):")
print(f"Shape: {df.shape}")
print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
print(f"Total sales: ${df['Sales'].sum():,}")
print("\nFirst 10 rows:")
print(df.head(10))
print("\nCategory distribution:")
print(df['Category'].value_counts())
print("\nSummary statistics:")
print(df['Sales'].describe())
```

```python
top_items = df.groupby('Item')['Sales'].sum().sort_values(ascending=False).head(100)
print(top_items)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Sales',data=df)
plt.show()
```

```python
# Trend
sns.lineplot(x='Date', y='Sales', data=df)
plt.show()
```

```python
# Distribution of data
sns.histplot(df['Sales'],bins=10)
plt.xlim(0,1000)
plt.show()
```

```python
#Box plot
sns.boxplot(x='Category', y='Sales', data=df)
plt.show()
```
