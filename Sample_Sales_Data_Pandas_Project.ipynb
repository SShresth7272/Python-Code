```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['sales_data_sample (1).csv']))
```

```python
df['Total'] = df['QUANTITYORDERED'] * df['PRICEEACH']
print(df)
df.head(5)
```

```python
df['YEAR_ID'] = pd.to_datetime(df['ORDERDATE']).dt.year
df['MONTH_ID'] = pd.to_datetime(df['ORDERDATE']).dt.month
display(df)
```

```python
total_sales = df.groupby('CITY')['SALES'].sum()
print(total_sales)
```

```python
most_popular = df.groupby('PRODUCTLINE')['SALES'].sum().idxmax()
highest_sales = df.groupby('PRODUCTLINE')['SALES'].sum().max()

print(f"Most popular product: {most_popular}")
print(f"Total sales: ${highest_sales:,.2f}")
```
