```python
#Filtering data based on conditions.
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 17, 50, 33, 29],
    'Salary': [70000, 48000, 110000, 95000, 65000]
}
df=pd.DataFrame(data)
df
```

```python
df['Age']
```

```python
# Filter for Age > 30
df_older= df[df['Age']>30]
df_older
```

```python
# Filter for Age > 25 and Salary > 95000
condition=(df['Age']>25)&(df['Salary']>95000)
df_filtered=df[condition]
print(df_filtered)
```

Group by and aggregate

```python
data = {
    'Department': ['Sales', 'Marketing', 'IT', 'HR', 'Sales', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Salary': [70000, 48000, 110000, 50000, 72000, 95000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Group by 'Department'
grouped= df.groupby('Department')
```

```python
# Calculate mean salary per department
mean_salary=grouped['Salary'].mean()

print(mean_salary)
```
